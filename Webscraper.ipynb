```python
from bs4 import BeautifulSoup
from selenium import webdriver
import json
import pandas as pd
import requests
```

```python
#configuring dataset display
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

#reviews ultimately displayed here
df = pd.DataFrame({'review': []})

#for Indeed in particular user reviews are stored in json format
script_list = []
company_name = input("Enter a company name: \n")
for i in range(0,140,20):
    url = (f'https://www.indeed.com/cmp/JPMorgan-Chase-&-Co-7555c073/reviews?start={i}')
    dr = webdriver.Chrome()
    dr.get(url)
    bs = BeautifulSoup(dr.page_source,"html")
    bs_all = bs.find_all("script", type="application/json")

    script_list.append(bs_all[1])

reviews_list = []
for script in script_list:
    res_dict = json.loads(script.contents[0])
    emp_reviews = res_dict['reviewsList']['items']
    
    for item in emp_reviews:
        review = item['text']['text']
        reviews_list.append(review)
        
final_rev_list = []
for review in reviews_list:
    final_rev_list.append(review.replace("\n", ""))

for review in final_rev_list:
    df = pd.concat([df, pd.DataFrame.from_records([{'review': review}])], ignore_index=True)

display(df)
```
